```python
import pandas as pd
```

```python
# Loading CSV file

inventory_data = pd.read_csv('/content/drive/MyDrive/Housing Project/inventory_data.csv')
majors_data = pd.read_csv('/content/drive/MyDrive/Housing Project/majors_data.csv')
occupancy_data = pd.read_csv('/content/drive/MyDrive/Housing Project/occupancy_data.csv')
persons_data = pd.read_csv('/content/drive/MyDrive/Housing Project/persons_data.csv')
```

```python
# Identify duplicate emails
duplicates = persons_data[persons_data.duplicated(subset=['email'], keep=False)]

# # Remove duplicates from the final_df
persons_data.drop_duplicates(subset=['email'], inplace=True)
```

```python
duplicates
```

```python
persons_data
```

```python
# Merging person_info with person_bed_info on personId
merged_df = pd.merge(persons_data, occupancy_data, on='personId', how='left')
```

```python
merged_df
```

```python
# Merging merged_df with building_info to get bedId
merged_df = pd.merge(merged_df, inventory_data[['buildingName', 'roomName', 'bedName', 'bedId']],
                     on=['buildingName', 'roomName', 'bedName'], how='left')
```

```python
merged_df
```

```python
# Selecting the required columns and renaming them
final_df = merged_df[['personId', 'firstName', 'lastName', 'email', 'dob', 'address', 'majors', 'bedId']].copy()
```

```python
final_df
```

```python
# Aggregate the display IDs for each major name into a list
majors_mapping = majors_data.groupby('name')['displayId'].apply(list).to_dict()
```

```python
majors_mapping
```

```python
# Function to replace majors with displayId lists
def replace_majors_with_displayIds(majors, mapping):
    if pd.isna(majors):
        return majors
    majors_list = majors.split(', ')
    display_ids = [', '.join(mapping.get(major.strip(), [major])) for major in majors_list]
    return '; '.join(display_ids)
```

```python
final_df['majors'] = final_df['majors'].apply(lambda x: replace_majors_with_displayIds(x, majors_mapping))
```

```python
final_df
```

```python
final_df.rename(columns={
    'firstName': 'name',
    'address': 'address1',
    'majors': 'majorIds'
}, inplace=True)
```

```python
# Concatenating first and last name into a full name
final_df['name'] = final_df['name'] + ' ' + merged_df['lastName']
final_df.drop(columns=['lastName'], inplace=True)
```

```python
final_df
```

```python
# Function to split address correctly
def split_address(address):
    if pd.isna(address):
        return pd.Series([None, None, None, None, None])
    parts = address.split(',')
    parts = [part.strip() for part in parts]
    if len(parts) == 4:
        return pd.Series([parts[0], parts[1], parts[2], parts[3], None])
    elif len(parts) == 3:
        return pd.Series([parts[0], None, parts[1], parts[2], None])
    elif len(parts) == 2:
        return pd.Series([parts[0], None, parts[1], None, None])
    elif len(parts) == 1:
        return pd.Series([parts[0], None, None, None, None])
    return pd.Series([None, None, None, None, None])

# Splitting the address into address1, address2, city, state, and zip
final_df[['address1', 'address2', 'city', 'state', 'zip']] = final_df['address1'].apply(split_address)
```

```python
final_df
```

```python
# Combine duplicates with invalid_rows
invalid_rows = pd.concat([duplicates, final_df[final_df['personId'].isnull() | final_df['email'].isnull()]])
```

```python
invalid_rows
```

```python
# Ensure the columns are in the required order
valid_rows = final_df[['personId', 'name', 'email', 'dob', 'address1', 'address2', 'city', 'state', 'zip', 'majorIds', 'bedId']]
invalid_rows = invalid_rows[['personId', 'name', 'email', 'dob', 'address1', 'address2', 'city', 'state', 'zip', 'majorIds', 'bedId']]
```

```python
# Convert columns to required data types for valid rows
valid_rows.loc[:, 'personId'] = valid_rows['personId'].astype(str)
valid_rows.loc[:, 'name'] = valid_rows['name'].astype(str)
valid_rows.loc[:, 'email'] = valid_rows['email'].astype(str)
valid_rows.loc[:, 'dob'] = pd.to_datetime(valid_rows['dob']).dt.strftime('%Y-%m-%d')
valid_rows.loc[:, 'address1'] = valid_rows['address1'].astype(str)
valid_rows.loc[:, 'address2'] = valid_rows['address2'].astype(str)
valid_rows.loc[:, 'city'] = valid_rows['city'].astype(str)
valid_rows.loc[:, 'state'] = valid_rows['state'].astype(str)
valid_rows.loc[:, 'zip'] = valid_rows['zip'].astype(str)
valid_rows.loc[:, 'majorIds'] = valid_rows['majorIds'].astype(str)
valid_rows.loc[:, 'bedId'] = valid_rows['bedId'].astype(str)

# Convert columns to required data types for invalid rows
invalid_rows.loc[:, 'personId'] = invalid_rows['personId'].astype(str)
invalid_rows.loc[:, 'name'] = invalid_rows['name'].astype(str)
invalid_rows.loc[:, 'email'] = invalid_rows['email'].astype(str)
invalid_rows.loc[:, 'dob'] = pd.to_datetime(invalid_rows['dob']).dt.strftime('%Y-%m-%d')
invalid_rows.loc[:, 'address1'] = invalid_rows['address1'].astype(str)
invalid_rows.loc[:, 'address2'] = invalid_rows['address2'].astype(str)
invalid_rows.loc[:, 'city'] = invalid_rows['city'].astype(str)
invalid_rows.loc[:, 'state'] = invalid_rows['state'].astype(str)
invalid_rows.loc[:, 'zip'] = invalid_rows['zip'].astype(str)
invalid_rows.loc[:, 'majorIds'] = invalid_rows['majorIds'].astype(str)
invalid_rows.loc[:, 'bedId'] = invalid_rows['bedId'].astype(str)
```
